```python
# ACC102 Mini Assignment - Track 2: GitHub Data Analysis Project
# Title: Financial Ratio Analysis for Three Listed Companies
# Student Name: Ruihan.Sun
# Student ID: 2469984
# Data Source: Official investor relations websites of Alibaba, Tencent, Pinduoduo
# Date of data access: April 24, 2026

# ---------------------------------------------------
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Set plot style 
plt.rcParams['font.sans-serif'] = ['DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
```

```python
#-----------------------------------------
# 1. Problem Definition & Intended User
# Analytical problem: Compare profitability and solvency of three internet companies
# Intended users: Finance students, investors, and business analysts
```

```python
# ---------------------------------------------------
# 2. Dataset Description
# Data collected manually from official company websites:
# - Alibaba: https://www.alibabagroup.com/en/investor-relations
# - Tencent: https://www.tencent.com/en-us/investor
# - Pinduoduo: https://ir.pinduoduo.com/
# Financial ratios include: Gross Margin, Net Margin, ROE, Debt-to-Asset, Current Ratio

data = {
    'Company': ['Alibaba', 'Tencent', 'Pinduoduo'],
    'Gross_Margin_Pct': [39.95, 56.20, 56.30],
    'Net_Margin_Pct': [8.91, 29.91, 23.01],
    'ROE_Pct': [8.23, 20.03, 27.29],
    'Debt_to_Asset_Pct': [20.58, 24.67, 34.15],
    'Current_Ratio': [1.55, 1.67, 2.20]
}

df = pd.DataFrame(data)
```

```python
# ---------------------------------------------------
# 3. Show dataset
print("=== Financial Dataset Collected from Official Websites ===")
print(df)
```

```python
# ---------------------------------------------------
# 4. Profitability Analysis Visualization
plt.figure(figsize=(10, 5))
companies = df['Company']
x = range(len(companies))
width = 0.25

plt.bar([i - width for i in x], df['Gross_Margin_Pct'], width, label='Gross Margin (%)')
plt.bar(x, df['Net_Margin_Pct'], width, label='Net Margin (%)')
plt.bar([i + width for i in x], df['ROE_Pct'], width, label='ROE (%)')

plt.title('Profitability Comparison of Three Companies')
plt.xlabel('Company')
plt.ylabel('Percentage (%)')
plt.xticks(x, companies)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# ---------------------------------------------------
# 5. Solvency Analysis Visualization
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(companies, df['Debt_to_Asset_Pct'], color='orange')
plt.title('Debt-to-Asset Ratio (%)')
plt.ylabel('%')
plt.grid(alpha=0.3)

plt.subplot(1, 2, 2)
plt.bar(companies, df['Current_Ratio'], color='green')
plt.title('Current Ratio')
plt.ylabel('Ratio')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# ---------------------------------------------------
# 6. Key Insights
print("\n=== Key Insights ===")
print("1. Tencent and Pinduoduo have higher gross margins than Alibaba.")
print("2. Pinduoduo shows the strongest ROE performance.")
print("3. Alibaba maintains the lowest debt-to-asset ratio.")
print("4. All companies demonstrate healthy current ratio levels.")
print("\nData source: Official investor relations websites, accessed April 2026.")
```
