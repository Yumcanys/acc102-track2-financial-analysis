```python
# ACC102 Mini Assignment - Track 2: GitHub Data Analysis Project
# Title: Financial Ratio Comparison Analysis of Listed Companies
# Analyst: [Ruihan.Sun]
# Data Source: Simulated financial data(for academic use only)
# Data Created: April 23rd, 2026
# ======================================================================
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# set plot style to avoid Chinese font issues(English-only for making)
plt.rcParams['font.sans-serif'] = ['DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# %%
# 1. Problem Definition and Objective
# This project analyzes and compares the profitability and solvency of three major.
# internet/tech companies: Alibaba, Tencent, and Pinduoduo.
# Users: Investors, finance students, and financial analysts.

# %%
# 2. Data Preparation
data = {
    'Company': ['Alibaba', 'Tencent', 'Pinduoduo'],
    'Gross_Margin_Pct': [39.95, 56.2, 56.3],
    'Net_Margin_Pct': [8.91, 29.91, 23.01],
    'ROE': [8.23, 20.03, 27.29],
    'Debt_to_Asset_Pct': [20.58, 24.67, 34.15],
    'Current_Ratio': [1.55, 1.67, 2.20] 
}

df = pd.DataFrame(data)

# %%
# 3. Show dataset
print("===== Financial Ratio Dataset =====")
display(df)

# %%
# 4. Profitability Analysis Visualization
plt.figure(figsize=(10, 5))
companies = df['Company']
x = range(len(companies))
width = 0.35

plt.bar([i - width/2 for i in x], df['Gross_Margin_Pct'], width, label='Gross Margin (%)')
plt.bar([i + width/2 for i in x], df['Net_Margin_Pct'], width, label='Net Margin (%)')

plt.title('Profitability Comparison: Gross Margin vs Net Margin')
plt.xlabel('Company')
plt.ylabel('Percentage (%)')
plt.xticks(x, companies)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()

# %%
# 5. Solvency Analysis Visualization
plt.figure(figsize=(12, 5))

# Debt to Asset Ratio
plt.subplot(1, 2, 1)
plt.bar(companies, df['Debt_to_Asset_Pct'], color='orange')
plt.title('Debt to Asset Ratio (%)')
plt.ylabel('%')
plt.grid(axis='y', alpha=0.3)

# Current Ratio
plt.subplot(1, 2, 2)
plt.bar(companies, df['Current_Ratio'], color='green')
plt.title('Current Ratio')
plt.ylabel('Ratio')
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# %%
# 6. Key Findings and Conclusion (Academic Style)
print("\n===== Key Insights and Conclusion =====")
print("1. Profitability: Pinduoduo has the highest gross margin and ROE, while Tencent leads in net margin.")
print("2. Solvency: Alibaba has the lowest debt ratio; Pinduoduo shows the highest current ratio.")
print("3. Overall: Pinduoduo leads in profitability and liquidity. Tencent has balanced performance. Alibaba is the most conservative.")
print("4. Limitation: Data is simulated; real annual report data is recommended for formal investment analysis.")
```
